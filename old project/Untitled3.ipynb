```python
import pandas as pd
import networkx as nx
import numpy as np
from pandas import DataFrame
```

```python
nodes = pd.read_csv('nodes.csv')
edges = pd.read_csv('edges.csv')
edges['x'] = edges['name'].map(dict(zip(nodes['id'],nodes['x'])))
edges['y'] = edges['name'].map(dict(zip(nodes['id'],nodes['y'])))
edges['r'] = edges['id']+'-->'+edges['target']

edges_dup = edges.copy()
temp = edges_dup['id'].copy()
edges_dup['id'] = edges_dup['target']
edges_dup['target'] = temp

pd.concat([edges,edges_dup]).to_csv('edges4.csv')
```
